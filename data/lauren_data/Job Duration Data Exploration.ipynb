```python
jobs = pandas.read_csv('../data.csv')
jobs = jobs[jobs.Duration_yr >= 0]
jobs = jobs[jobs.EndDate != 'Present']
```

```python
jobGroups = jobs.groupby('JobType')
summaryTable = jobGroups['Duration_yr'].agg([np.sum, np.mean, np.std, np.median, len]).sort('median', ascending=False)
print summaryTable
```

```python
jobscur = pandas.read_csv('../datacur.csv')
jobscur = jobscur[jobscur.Duration_yr >= 0]
```

```python
jobscur.head()
```

```python
jobscurGroups = jobscur.groupby('JobType')
summaryTableCur = jobscurGroups['Duration_yr'].agg([np.sum, np.mean, np.std, np.median, len]).sort('median', ascending=False)
print summaryTableCur
```

```python
summaryTableCur = summaryTableCur.drop(['Student','Other'])
print summaryTableCur
```

```python
# comment

jobFunctions = pandas.unique(summaryTable.index.ravel()).tolist()
currentMedians = []
currentNumberPoints = []
currentStds = []
for i in jobFunctions:
    currentMedians.append(summaryTableCur['median'][i])
    currentStds.append(summaryTableCur['std'][i])
    currentNumberPoints.append('# points: ' + str(summaryTableCur['len'][i]))

traces= None
traces = [
    Bar(
        x = jobFunctions,
        y = numpy.array(summaryTable['median']),
        name = 'past jobs',
        text = ["# points: "+str(i) for i in (summaryTable['len'].tolist())],
        error_y=ErrorY(
            type='data',
            array=numpy.array(summaryTable['std']),
            visible=False
        )
    ), 
    Bar(
        x = jobFunctions,
        y = currentMedians,
        name = 'current jobs',
        text = currentNumberPoints,
        error_y=ErrorY(
            type='data',
            array=currentStds,
            visible=False
        )
    )
]

layout = Layout(
    barmode='group',
    font=Font(
        family='Raleway, sans-serif',
        size=18,
    ),
    xaxis=XAxis(
        tickangle=-45
    ),
    yaxis=YAxis(
        title='job duration (years)',
        #zeroline=False,
        gridwidth=2
    ),
)

# Make Data object made up of 120 Box object 
plotdata = Data(traces)

# Make figure object
fig = Figure(data=plotdata, layout=layout)

# (@) Send to Plotly and show plot in notebook
py.iplot(fig)
```
